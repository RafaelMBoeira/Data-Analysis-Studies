### With Matplotlib

```python
x = np.linspace(-10, 10, 100)
```

```python
def func(x):
  return np.power(x, 2) - 10
```

```python
y = func(x)
```

```python
plt.plot(x, y, color='red', linewidth=3, linestyle='--')
plt.title('y = x')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
plt.scatter(x, y, color='red', marker='v')
plt.title('Ice Cream Ratings')
plt.xlabel('Flavor Rating')
plt.ylabel('Overall Rating')
plt.show()
```

```python
fig = plt.figure(figsize=(6, 5))

axes_a = fig.add_axes([0, 0, 1, 1])
axes_b = fig.add_axes([0.4, 0.15, 0.5, 0.3])

axes_a.plot(x, y, color='red', ls='--')
axes_a.set_xlabel('x')
axes_a.set_ylabel('y')
axes_a.set_title('y = x - 10')

axes_a.plot(x, np.pow(x, 3), color='green')
axes_a.legend(['x-10', 'x'], loc=0)

axes_b.plot(y, x, ls=':')
axes_b.set_xlabel('y')
axes_b.set_ylabel('x')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(6,5))

axes[0].plot(x, y)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('y = x - 10')

axes[1].plot(y, x)
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')

plt.tight_layout()
```

# Cleaning Data

```python
df = pd.read_excel("/content/Customer Call List.xlsx")
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns=['Not_Useful_Column'], inplace=True)
```

```python
df.fillna('', inplace=True)
```

```python
df['Paying Customer'].unique()
```

```python
import re
```

```python
df['Last_Name'] = df['Last_Name'].apply(lambda x: re.sub('[^a-zA-Z]', '', str(x)))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: re.sub('[^0-9]', '', str(x)))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df[['Street Address', 'State', 'Zipcode']] = df['Address'].str.split(',', expand=True)
df.drop(columns=['Address'], inplace=True)
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df.fillna('', inplace=True)
```

```python
df['Zipcode'] = df['Zipcode'].apply(lambda x: str(x))
df['State'] = df['State'].apply(lambda x: str(x))
```

```python
df.replace('N/a', '', inplace=True)
df.replace('--', '', inplace=True)
```

```python
df
```

```python
for i in df.index:
  if df.loc[i, 'Do_Not_Contact'] == 'Y':
    df.drop(i, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```
